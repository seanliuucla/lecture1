```python
df_song = df.loc[df['Track Name']=='Chantaje']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
x_axis = df_song['Date']
y_axis = df_song['Streams']
plt.bar(x_axis, y_axis)
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Streams')
plt.ylim(ymin=10000)

plt.title("Chantaje Streams Plot")
plt.grid()
plt.show()
```

Observation: There appears to be a weekly pattern for the streams of Chantaje. 20217/01/01 was a Sunday. It seems that higher streams happened on Saturdays, folled by Sundays.
